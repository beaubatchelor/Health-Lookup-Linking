```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#import world happiness report, based on data from Gallup World Poll
happiness_df = pd.read_csv("../Data/2016.csv")
happiness_df = happiness_df.rename(columns={"Happiness.Score" : "Happiness Score"})
```

```python
#clean happiness_df to only give columns with "Country", "Region", and "Happiness Score"
happiness_df=happiness_df[["Country", "Region", "Happiness Score"]]
happiness_df.head()
```

```python
#import Life Expectancy Data, data from CIA.gov
expected_life_df =  pd.read_csv("../Data/LifeExpectancy.csv")
expected_life_df.head()
```

```python
#clean expected_life_df
for index, row in expected_life_df.iterrows():
    country = row['Column2'].rstrip()
    expected_life_df.iloc[index, 1] = country
    
expected_life_df = expected_life_df.drop(columns=["Column1"])
expected_life_df=expected_life_df.rename(columns={"Column2":"Country", "Column3":"Life Expectancy"})
expected_life_df.head()
```

```python
#merge dataframes
merged_happiness_ELife = pd.merge(happiness_df, expected_life_df, how = "left", on="Country")
merged_happiness_ELife.head()
```

```python

sns.lmplot(x="Life Expectancy", y="Happiness Score", data=merged_happiness_ELife, ci=False, truncate=True)
plt.xlabel("Life Expectancy at Birth")
plt.title("Happiness vs Life Expectancy")
plt.ylim(2, 8)
```

#### There is a positive correlation between life expectancy and happiness. As life expectancy increases, happiness increases. However, the difference between the observed values and the linear model prediction is large, so .

```python
#Categorize Region
sns.scatterplot(x="Life Expectancy", y="Happiness Score", hue="Region", data=merged_happiness_ELife)
plt.legend(bbox_to_anchor= (1, 1))
plt.xlabel("Life Expectancy at Birth")
plt.title("Happiness vs Life Expectancy by Region")
```

#### After color-coding the country's region, we can see that, Sub-Saharan Africa is the least happy region, and Western Europe is the happiest region.
